```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Sample data
data1 = pd.DataFrame({'a': [12.34, -145.67]})
data2 = pd.DataFrame({'a': [12.34, 15.67]})
data3 = pd.DataFrame({'a': ['AA RT O', 'AS/ER NB']})
data4 = pd.DataFrame({'a': [pd.to_datetime('2004-10-03'), pd.to_datetime('2004-10-15')]})
data5 = pd.DataFrame({'a': [13, 45, 56, 67, 87, 256]})
data6 = pd.DataFrame({'a': ['125678', '342639', '563843', '655454']})
data7 = pd.DataFrame({'a': [12.34, 15.67, -12.45, -14.66]})

# Absolute Function
data1['x'] = data1['a'].abs()
print(data1)

# Ceil Function
data2['x'] = np.ceil(data2['a'])
print(data2)

# Floor Function
data2['x'] = np.floor(data2['a'])
print(data2)

# Compress Function
data3['x'] = data3['a'].str.replace(' ', '')
data3['y'] = data3['a'].str.replace('/', '')
data3['z'] = data3['a'].str.replace('/', '').str.replace(' ', '')
print(data3)

# Date Function
data4 = pd.DataFrame({'tday': [pd.Timestamp.now().date()]})
print(data4)

# Datetime Function
data4 = pd.DataFrame({'tday': [pd.Timestamp.now()]})
print(data4)

# Datepart Function
data4['x'] = data4['tday'].dt.date
print(data4)

# Day Function
data4['x'] = data4['tday'].dt.day
print(data4)

# DHMS Function
data5 = pd.DataFrame({'dt': [pd.Timestamp(2004, 10, 1, 11, 34, 12)]})
print(data5)

# DIF Function
data5['x'] = data5['a'].diff()
data5['x1'] = data5['a'].diff(1)
data5['x2'] = data5['a'].diff(2)
print(data5)

# Exponential Function
data5 = pd.DataFrame({'x': [np.exp(1)], 'y': [np.exp(0)]})
print(data5)

# HMS Function
data5 = pd.DataFrame({'tt': [pd.Timestamp(2004, 10, 1, 11, 34, 12).time()]})
print(data5)

# Hour Function
data5['hh'] = data5['tt'].apply(lambda x: x.hour)
print(data5)

# Input Function
data6['x'] = pd.to_numeric(data6['a'])
data6['z'] = data6['a'].astype(int)
print(data6)

# INT Function
data7['x'] = data7['a'].astype(int)
print(data7)

# INTCK Function
data8 = pd.DataFrame({
    'dd': [(datetime(1999, 12, 31) - datetime(1999, 1, 1)).days],
    'mm': [(datetime(1999, 12, 31).year - datetime(1999, 1, 1).year) * 12 + (datetime(1999, 12, 31).month - datetime(1999, 1, 1).month)],
    'ww': [(datetime(1999, 12, 31) - datetime(1999, 1, 1)).days // 7],
    'qq': [(datetime(1999, 12, 31).year - datetime(1999, 1, 1).year) * 4 + (datetime(1999, 12, 31).month - 1) // 3 - (datetime(1999, 1, 1).month - 1) // 3],
    'yy': [datetime(1999, 12, 31).year - datetime(1999, 1, 1).year]
})
print(data8)

# INTNX Function
data9 = pd.DataFrame({
    'dd': [datetime(1999, 12, 31) - timedelta(days=364)],
    'mm': [datetime(1999, 12, 31) - timedelta(days=30*11)],
    'yy': [datetime(1999, 12, 31)]
})
print(data9)

# Soundex Function
# Using jellyfish library for Soundex
import jellyfish

data10 = pd.DataFrame({'ordered_by': ['Rob', 'Bob'], 'ordered_for': ['Tom', 'Bom']})
data10['sound1'] = data10['ordered_by'].apply(jellyfish.soundex)
data10['sound2'] = data10['ordered_for'].apply(jellyfish.soundex)
print(data10)
```
